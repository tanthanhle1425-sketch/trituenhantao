```python
#Cau13
def merge_arrays_with_sum_and_remainder(arr1, arr2):
    """
    Trộn 2 mảng một chiều thành 1 mảng một chiều.
    Mỗi phần tử của mảng mới là tổng của 2 phần tử tương ứng từ 2 mảng cho trước
    cho đến khi một trong hai mảng hết phần tử.
    Các phần tử còn lại của mảng dài hơn sẽ được đưa vào mảng mới.

    Args:
        arr1 (list): Mảng thứ nhất.
        arr2 (list): Mảng thứ hai.

    Returns:
        list: Mảng kết quả đã trộn.
    """
    result = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    # Tính tổng các phần tử tương ứng cho đến hết mảng ngắn hơn
    for i in range(min_len):
        result.append(arr1[i] + arr2[i])

    # Thêm các phần tử còn lại của mảng dài hơn vào kết quả
    if len1 > len2:
        result.extend(arr1[min_len:])
    elif len2 > len1:
        result.extend(arr2[min_len:])

    return result

# Ví dụ sử dụng dựa trên mô tả
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
merged_result = merge_arrays_with_sum_and_remainder(a, b)
print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả (theo mô tả): {merged_result}")

# Các trường hợp kiểm tra khác
# Trường hợp mảng arr1 dài hơn
c = [10, 20, 30, 40, 50]
d = [1, 2, 3]
merged_result_c_d = merge_arrays_with_sum_and_remainder(c, d)
print(f"\nMảng c: {c}")
print(f"Mảng d: {d}")
print(f"Mảng kết quả (c, d): {merged_result_c_d}")

# Trường hợp hai mảng có cùng độ dài
e = [1, 2, 3]
f = [4, 5, 6]
merged_result_e_f = merge_arrays_with_sum_and_remainder(e, f)
print(f"\nMảng e: {e}")
print(f"Mảng f: {f}")
print(f"Mảng kết quả (e, f): {merged_result_e_f}")

# Trường hợp có mảng rỗng
g = []
h = [1, 2, 3]
merged_result_g_h = merge_arrays_with_sum_and_remainder(g, h)
print(f"\nMảng g: {g}")
print(f"Mảng h: {h}")
print(f"Mảng kết quả (g, h): {merged_result_g_h}")
```

```python
#cau14
import numpy as np

def create_random_2d_array(m, n, min_val=0, max_val=100):
    """
    Tạo một mảng 2 chiều có m dòng, n cột chứa các số nguyên ngẫu nhiên.

    Args:
        m (int): Số lượng dòng.
        n (int): Số lượng cột.
        min_val (int): Giá trị nhỏ nhất của số nguyên ngẫu nhiên (bao gồm).
        max_val (int): Giá trị lớn nhất của số nguyên ngẫu nhiên (không bao gồm).

    Returns:
        np.ndarray: Mảng 2 chiều chứa các số nguyên ngẫu nhiên.
    """
    if m <= 0 or n <= 0:
        raise ValueError("Số dòng (m) và số cột (n) phải là số nguyên dương.")

    # Sử dụng np.random.randint để tạo mảng số nguyên ngẫu nhiên
    random_array = np.random.randint(min_val, max_val, size=(m, n))
    return random_array

print("Hàm `create_random_2d_array` đã được định nghĩa.")
# Ví dụ sử dụng hàm với m và n đã cho trong kernel (m_large, n_large)
# Giả sử m_large và n_large đã được định nghĩa từ trước, ví dụ: m_large = 5, n_large = 5

# Lấy giá trị m và n từ kernel state
m = m_large # m_large = 5 từ kernel state
n = n_large # n_large = 5 từ kernel state

print(f"Tạo mảng 2 chiều với {m} dòng và {n} cột:")
a = create_random_2d_array(m, n, min_val=1, max_val=50) # Ví dụ tạo số từ 1 đến 49
print(a)
```

```python
#Cau14: a.Xuất các phần tử thuộc dòng k

# Kiểm tra xem mảng 'a' có tồn tại trong kernel không
if 'a' not in locals() and 'a' not in globals():
    print("Mảng 'a' chưa được tạo. Vui lòng chạy lại các ô trước để tạo mảng 'a'.")
else:
    print(f"Mảng 'a' hiện tại:\n{a}")
    num_rows = a.shape[0]
    while True:
        try:
            k = int(input(f"Nhập chỉ số dòng k (từ 0 đến {num_rows - 1}) để xuất các phần tử: "))
            if 0 <= k < num_rows:
                break
            else:
                print(f"Chỉ số dòng k không hợp lệ. Vui lòng nhập số nguyên từ 0 đến {num_rows - 1}.")
        except ValueError:
            print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
    # Xuất các phần tử thuộc dòng k
    row_k = a[k, :]
    print(f"Các phần tử thuộc dòng {k} là: {row_k}")
```

```python
#Cau14: b. Xuất các phần tử thuộc cột k

# Kiểm tra xem mảng 'a' có tồn tại trong kernel không
if 'a' not in locals() and 'a' not in globals():
    print("Mảng 'a' chưa được tạo. Vui lòng chạy lại các ô trước để tạo mảng 'a'.")
else:
    print(f"Mảng 'a' hiện tại:\n{a}")
    num_cols = a.shape[1]

    while True:
        try:
            k = int(input(f"Nhập chỉ số cột k (từ 0 đến {num_cols - 1}) để xuất các phần tử: "))
            if 0 <= k < num_cols:
                break
            else:
                print(f"Chỉ số cột k không hợp lệ. Vui lòng nhập số nguyên từ 0 đến {num_cols - 1}.")
        except ValueError:
            print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

    # Xuất các phần tử thuộc cột k
    col_k = a[:, k]
    print(f"Các phần tử thuộc cột {k} là: {col_k}")
```

```python
#Cau14: c. Tìm dòng có tổng lớn nhất và d. Tìm cột có tích nhỏ nhất

# Kiểm tra xem mảng 'a' có tồn tại trong kernel không
if 'a' not in locals() and 'a' not in globals():
    print("Mảng 'a' chưa được tạo. Vui lòng chạy lại các ô trước để tạo mảng 'a'.")
else:
    print(f"Mảng 'a' hiện tại:\n{a}")

    # c. Tìm dòng có tổng lớn nhất
    row_sums = np.sum(a, axis=1) # Tính tổng từng dòng
    max_sum_row_index = np.argmax(row_sums) # Tìm chỉ số của dòng có tổng lớn nhất
    max_sum_row = a[max_sum_row_index, :]
    print(f"\nDòng có tổng lớn nhất là dòng {max_sum_row_index} với tổng = {row_sums[max_sum_row_index]}: {max_sum_row}")

    # d. Tìm cột có tích nhỏ nhất
    # Sử dụng np.prod để tránh tràn số với các số lớn, nếu mảng chứa các giá trị rất lớn.
    # Tuy nhiên, nếu mảng chứa 0, tích sẽ là 0, điều này có thể ảnh hưởng đến kết quả.
    # Đảm bảo các giá trị trong mảng không âm để tích có ý nghĩa khi so sánh nhỏ nhất.
    column_products = np.prod(a, axis=0) # Tính tích từng cột
    min_prod_col_index = np.argmin(column_products) # Tìm chỉ số của cột có tích nhỏ nhất
    min_prod_col = a[:, min_prod_col_index]
    print(f"Cột có tích nhỏ nhất là cột {min_prod_col_index} với tích = {column_products[min_prod_col_index]}: {min_prod_col}")
```

```python
#Cau14: f. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a

# Kiểm tra xem mảng 'a' có tồn tại trong kernel không
if 'a' not in locals() and 'a' not in globals():
    print("Mảng 'a' chưa được tạo. Vui lòng chạy lại các ô trước để tạo mảng 'a'.")
else:
    print(f"Mảng 'a' hiện tại:\n{a}")

    # Lấy các dòng có chỉ số lẻ (1, 3, 5, ...)
    odd_rows = a[1::2, :]
    print(f"\nCác dòng lẻ:\n{odd_rows}")

    # Lọc ra các phần tử chẵn từ các dòng lẻ
    even_elements_in_odd_rows = odd_rows[odd_rows % 2 == 0]

    if even_elements_in_odd_rows.size > 0:
        average = np.mean(even_elements_in_odd_rows)
        print(f"Các phần tử chẵn thuộc dòng lẻ: {even_elements_in_odd_rows}")
        print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ là: {average:.2f}")
    else:
        print("Không có phần tử chẵn nào thuộc dòng lẻ.")
```

```python
#Cau14: g. Tính trung bình cộng các phần tử thuộc biên và h. Tính tích các phần tử không thuộc biên

# Kiểm tra xem mảng 'a' có tồn tại trong kernel không
if 'a' not in locals() and 'a' not in globals():
    print("Mảng 'a' chưa được tạo. Vui lòng chạy lại các ô trước để tạo mảng 'a'.")
else:
    print(f"Mảng 'a' hiện tại:\n{a}")

    num_rows, num_cols = a.shape

    # g. Tính trung bình cộng các phần tử thuộc biên
    if num_rows == 0 or num_cols == 0:
        print("\nMảng rỗng, không có phần tử biên.")
        average_border = None
    else:
        # Tạo một mask boolean để đánh dấu các phần tử biên
        border_mask = np.full(a.shape, False, dtype=bool)
        border_mask[0, :] = True  # Hàng đầu tiên
        border_mask[-1, :] = True # Hàng cuối cùng
        border_mask[:, 0] = True  # Cột đầu tiên
        border_mask[:, -1] = True # Cột cuối cùng

        border_elements = a[border_mask]

        if border_elements.size > 0:
            average_border = np.mean(border_elements)
            print(f"\nCác phần tử thuộc biên là: {border_elements}")
            print(f"Trung bình cộng các phần tử thuộc biên là: {average_border:.2f}")
        else:
            print("\nKhông có phần tử biên (ví dụ: mảng 1x1).")
            average_border = None

    # h. Tính tích các phần tử không thuộc biên
    # Các phần tử không thuộc biên là a[1:-1, 1:-1]
    # Slicing a[1:-1, 1:-1] sẽ trả về mảng rỗng nếu num_rows < 3 hoặc num_cols < 3
    if num_rows < 3 or num_cols < 3:
        print("\nKích thước mảng quá nhỏ, không có phần tử không thuộc biên.")
        product_non_border = None
    else:
        non_border_elements = a[1:-1, 1:-1]
        if non_border_elements.size > 0:
            product_non_border = np.prod(non_border_elements)
            print(f"\nCác phần tử không thuộc biên là:\n{non_border_elements}")
            print(f"Tích các phần tử không thuộc biên là: {product_non_border}")
        else:
            print("\nKhông có phần tử không thuộc biên (có thể do kích thước mảng không đủ lớn sau khi cắt).")
            product_non_border = None
```

```python
#cau15
class SINHVIEN:
    def __init__(self, ma_sinh_vien, ten, nam_sinh, diem_trung_binh):
        # 1. Mã sinh viên là chuỗi có 10 ký tự
        if not isinstance(ma_sinh_vien, str) or len(ma_sinh_vien) != 10:
            raise ValueError("Mã sinh viên phải là chuỗi có đúng 10 ký tự.")
        self.ma_sinh_vien = ma_sinh_vien
        # 2. Tên là chuỗi tối đa 20 ký tự
        if not isinstance(ten, str) or len(ten) > 20:
            raise ValueError("Tên phải là chuỗi và không quá 20 ký tự.")
        self.ten = ten
        # 3. Năm sinh là một số nguyên
        if not isinstance(nam_sinh, int):
            raise ValueError("Năm sinh phải là một số nguyên.")
        self.nam_sinh = nam_sinh
        # 4. Điểm trung bình là một số thực
        if not isinstance(diem_trung_binh, (int, float)):
            raise ValueError("Điểm trung bình phải là một số thực.")
        self.diem_trung_binh = float(diem_trung_binh)
    def __str__(self):
        return (f"Mã sinh viên: {self.ma_sinh_vien}\n" +
                f"Tên: {self.ten}\n" +
                f"Năm sinh: {self.nam_sinh}\n" +
                f"Điểm trung bình: {self.diem_trung_binh:.2f}")

print("Cấu trúc SINHVIEN đã được định nghĩa.")

# Ví dụ sử dụng cấu trúc SINHVIEN
try:
    sv1 = SINHVIEN("SV00000001", "Nguyễn Văn A", 2002, 8.5)
    print("\n--- Thông tin sinh viên 1 ---")
    print(sv1)

    sv2 = SINHVIEN("SV00000002", "Trần Thị Bích Vân", 2003, 7.85)
    print("\n--- Thông tin sinh viên 2 ---")
    print(sv2)

    # Ví dụ với dữ liệu không hợp lệ (mã sinh viên sai độ dài)
    # sv_error_ma = SINHVIEN("SV001", "Lỗi Mã", 2000, 6.0)
    # print(sv_error_ma)

    # Ví dụ với dữ liệu không hợp lệ (tên quá dài)
    # sv_error_ten = SINHVIEN("SV00000003", "Phạm Minh Khôi An Ninh", 2001, 9.2)
    # print(sv_error_ten)

except ValueError as e:
    print(f"\nLỗi khi tạo sinh viên: {e}")
```

```python
#Cau15: Viết hàm đếm sinh viên đủ điều kiện lên lớp

def dem_sinh_vien_du_dieu_kien(danh_sach_sinh_vien):
    """
    Đếm số lượng sinh viên đủ điều kiện lên lớp (điểm trung bình >= 5).

    Args:
        danh_sach_sinh_vien (list): Danh sách các đối tượng SINHVIEN.

    Returns:
        int: Số lượng sinh viên đủ điều kiện.
    """
    so_luong_du_dieu_kien = 0
    for sv in danh_sach_sinh_vien:
        if sv.diem_trung_binh >= 5.0:
            so_luong_du_dieu_kien += 1
    return so_luong_du_dieu_kien

# Tạo một danh sách sinh viên mẫu để kiểm tra
try:
    ds_sv = [
        SINHVIEN("SV00000001", "Nguyễn Văn A", 2002, 8.5),
        SINHVIEN("SV00000002", "Trần Thị Bích Vân", 2003, 4.8),
        SINHVIEN("SV00000003", "Phan Văn C", 2004, 6.2),
        SINHVIEN("SV00000004", "Phạm Thị D", 2005, 9.1),
        SINHVIEN("SV00000005", "Hoàng Minh E", 2000, 3.5),
        SINHVIEN("SV00000006", "Phan Thị Lan", 2003, 7.0)
    ]
    print("\n--- Danh sách sinh viên mẫu đã tạo ---")
    for sv in ds_sv:
        print(f"  - {sv.ten} (Điểm TB: {sv.diem_trung_binh})")

    # Sử dụng hàm để đếm sinh viên đủ điều kiện
    so_luong_dat = dem_sinh_vien_du_dieu_kien(ds_sv)
    print(f"\nSố lượng sinh viên đủ điều kiện lên lớp: {so_luong_dat}")

except ValueError as e:
    print(f"\nLỗi khi tạo danh sách sinh viên: {e}")
```
